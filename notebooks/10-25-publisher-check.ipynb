```python
import pandas as pd
DATA_PATH = '../data/'
DATA_INTERIM_PATH = DATA_PATH + 'interim/'
train = pd.read_csv(DATA_INTERIM_PATH + 'train_c.csv')
```

```python
import pickle

UTIL_PATH = '../util/'
pickle_in = open(UTIL_PATH + "tld.pickle","rb")
tlds = pickle.load(pickle_in)
```

```python
train_pubs = set(train['domain'].values)
all_pubs = set(tlds.keys())
```
